**Readiness Score Function :**

```python
def readiness_score(attendance, diary, tasks, assignment, lab):
    score = (attendance*0.3) + (diary*2) + (tasks*3) + (assignment*0.2) + (lab*0.2)
    return round(score,2)

df["Readiness Score"] = df.apply(
    lambda x: readiness_score(
        x["attendance_pct"],
        x["diary_entries"],
        x["tasks_completed"],
        x["assignment_score"],
        x["lab_score"]
    ),
    axis=1
)

df.head()
```

###Markdown Cell

This function is used to calculate the readiness score for each student.

**Classification :**

```python
bands=[]

for score in df["Readiness Score"]:
    if score>=90:
        bands.append("Excellent")
    elif score>=70:
        bands.append("Strong")
    elif score>=50:
        bands.append("Developing")
    else:
        bands.append("Support")

df["Performance Band"]=bands

df
```

###Markdown Cell

Students are classified as Support, Developing, Strong, or Excellent based on their readiness score.

**Analytics :**

Department Average :

```python
print("Department Average")
print(df.groupby("department")["Readiness Score"].mean())
```

Top Performer :

```python
print("\nTop Performer")
print(df.loc[df["Readiness Score"].idxmax()])
```

Students Needing Support :

```python
print("\nStudents Needing Support")
print(df[df["Performance Band"]=="Support"])
```
